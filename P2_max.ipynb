```python
import pandas as pd

import json

with open('/Users/MaxAlex/Desktop/climate_videos_new.json', 'r') as f:
    data = f.readlines()

json_data = [json.loads(line) for line in data]

df = pd.DataFrame(json_data)
# Supposons que vous ayez un DataFrame 'df' contenant les données des vidéos avec les colonnes 'title', 'tags', 'description' et 'categories'

# Votre liste de keywords et leurs scores (comme mentionné précédemment)
# Par exemple, utilisons le dictionnaire 'scores' pour les mots-clés

# Dictionnaire de scores spécifique à chaque catégorie
category_scores = {
    'Travel & Events': 2,
    'Science & Technology': 1,
    'Education': 1,
    'Pets & Animals': 3,
    'People & Blogs': 2,
    'Comedy': 3,
    'Entertainment': 3,
    'News & Politics': 1,
    'Howto & Style': 3,
    'Sports': 3,
    'Autos & Vehicles': 3,
    'Music': 3,
    'Gaming': 3,
    'Film & Animation': 2,
    'Nonprofits & Activism': 1,
    '':3
}


# Fonction pour compter le nombre de points par catégorie pour chaque vidéo
def count_points(row):
    points = 0
    category = row['categories']
    
    # Vérifier chaque mot-clé et son score associé à la catégorie de la vidéo
    for keyword, score in scores.items():
        keyword_lower = keyword.lower()
        if keyword_lower in row['title'].lower():
            points += score * 3  # Multiplier le score par 3 si le mot-clé est dans le titre
        elif keyword_lower in row['tags'].lower():
            points += score * 2  # Multiplier le score par 2 si le mot-clé est dans les tags
        elif keyword_lower in row['description'].lower():
            points += score  # Ajouter 1 point si le mot-clé est dans la description
    
    return points > category_scores[category]  # Garder la vidéo si le score est supérieur ou égal au score de la catégorie

# Appliquer la fonction pour filtrer les vidéos selon les critères de score par catégorie
filtered_videos = df[df.apply(count_points, axis=1)]

# Afficher le DataFrame contenant les vidéos filtrées
print(filtered_videos)
```
