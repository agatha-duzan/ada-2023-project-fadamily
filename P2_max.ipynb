```python
# our keywords list
keywords = ['climate change', 'global warming', 'greenhouse gases', 'rising sea levels',
    'extreme weather', 'temperature increase', 'climate action', 
    'carbon emissions', 'renewable energy', 'sustainable living',
    'climate science', 'climate crisis', 'climate adaptation',
    'natural disaster', 'extreme heat', 'melting ice cap', 'sea level rise',
    'biodiversity loss', 'deforestation', 'ocean acidification', 'climate policy', 
    'environmental policy',
    'disaster recovery', 'climate refugees',
    'carbon footprint', 'sustainable development', 'green technology',
    'renewable resource', 'eco-friendly',
    'ecosystem disruption', 'impact on climate',
    'paris agreement', 'climate awareness',
    'environmental justice', 'clean energy', 'zero carbon',
    'green infrastructure', 'ozone layer',
    'pollution', 'water scarcity', 'climate education'] 

# Counting the number of keywords in the description
def count_keywords(description):
    count = 0
    for keyword in keywords:
        if keyword.lower() in description.lower():
            count += 1
    return count


# Filter the dataframe by the number of key words
filtered_df_2 = df[df['description'].apply(count_keywords) >= 2]
filtered_df_3 = df[df['description'].apply(count_keywords) >= 3]
filtered_df_4 = df[df['description'].apply(count_keywords) >= 4]


filtered_df_2.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# filtered_dfs_list est une liste de DataFrames filtrés
filtered_dfs_list = [df, filtered_df_2, filtered_df_3, filtered_df_4]
colors = ['skyblue', 'salmon', 'green', 'orange']

# Création d'un index global de toutes les catégories à partir de tous les DataFrames
categories = set()
for filtered_df in filtered_dfs_list:
    categories.update(filtered_df['categories'].unique())

categories.discard('')  # Retirer les catégories vides ('')

# Création d'une liste des catégories triées par ordre décroissant selon le nombre d'occurrences
category_counts = {category: sum(df['categories'].eq(category).sum() for df in filtered_dfs_list) for category in categories}
sorted_categories = sorted(categories, key=category_counts.get, reverse=True)

plt.figure(figsize=(12, 8))  # Réglage de la taille du graphique

# Largeur des barres pour chaque catégorie
bar_width = 0.15

# Création d'une liste de positions pour les barres pour chaque catégorie
bar_positions = np.arange(len(sorted_categories))

# Tracer les barres pour chaque catégorie triée et chaque DataFrame filtré
for i, filtered_df in enumerate(filtered_dfs_list):
    category_counts = filtered_df['categories'].value_counts()

    # Créer un dictionnaire pour les comptages de chaque catégorie, remplir les valeurs manquantes avec 0
    category_counts_dict = category_counts.to_dict()
    category_counts_full = {category: category_counts_dict.get(category, 0) for category in sorted_categories}

    # Décaler la position des barres pour chaque DataFrame
    plt.bar(bar_positions + i * bar_width, [category_counts_full[category] for category in sorted_categories], width=bar_width, color=colors[i], label=f'Filtre {i+1}')

plt.title('Occurrences des catégories pour chaque DataFrame (triées par ordre décroissant)')  # Titre du graphique
plt.xlabel('Catégories')  # Étiquette de l'axe x
plt.ylabel('Nombre d\'occurrences')  # Étiquette de l'axe y
plt.xticks(bar_positions + (len(filtered_dfs_list) - 1) * bar_width / 2, sorted_categories, rotation=45)  # Définir les étiquettes de l'axe x
plt.legend()  # Affichage de la légende
plt.tight_layout()  # Ajustement de la disposition pour éviter la superposition
plt.show()  # Affichage du graphique
```

```python
count_by_keyword = {keyword: 0 for keyword in keywords}

# Parcourir chaque description de vidéo et compter les occurrences des mots-clés
for description in df['description']:
    for keyword in keywords:
        if keyword.lower() in description.lower():
            count_by_keyword[keyword] += 1

# Créer une liste de nombres de vidéos pour chaque mot-clé
video_counts = [count for count in count_by_keyword.values()]

# Créer un graphique à barres pour afficher le nombre de vidéos par mot-clé
plt.figure(figsize=(12, 8))
plt.barh(keywords, video_counts, color='skyblue')
plt.xlabel('Nombre de vidéos')
plt.title('Nombre total de vidéos par mot-clé')
plt.gca().invert_yaxis()  # Inverser l'axe y pour afficher les mots-clés en ordre décroissant
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

dataframes = [df, filtered_df_2, filtered_df_3, filtered_df_4] 


count_by_keyword = {keyword: [0] * len(dataframes) for keyword in keywords}

# Parcourir chaque DataFrame pour compter les occurrences des mots-clés dans leurs descriptions
for idx, dataframe in enumerate(dataframes):
    for description in dataframe['description']:
        for keyword in keywords:
            if keyword.lower() in description.lower():
                count_by_keyword[keyword][idx] += 1

# Créer une liste de nombres de vidéos pour chaque mot-clé et chaque dataset
video_counts = [[count_by_keyword[keyword][i] for keyword in keywords] for i in range(len(dataframes))]

# Créer un graphique à barres groupées pour chaque mot-clé et chaque dataset
num_datasets = len(dataframes)
bar_width = 0.2
index = np.arange(len(keywords))

plt.figure(figsize=(20, 8))

for i in range(num_datasets):
    plt.bar(index + i * bar_width, video_counts[i], bar_width, label=f'Dataset {i+1}')

plt.xlabel('Mots-clés')
plt.ylabel('Nombre de vidéos')
plt.title('Nombre de vidéos par mot-clé et par dataset')
plt.xticks(index + bar_width * (num_datasets - 1) / 2, keywords, rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
